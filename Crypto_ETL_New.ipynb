```python
# Dependencies
import requests
from pprint import pprint
from config import api_key
import pandas as pd
from pprint import pprint
```

```python
url = f"https://min-api.cryptocompare.com/data/v2/histohour?fsym=BTC&tsym=USD&limit=2000&toTs=1635876000&api_key={api_key}"
r = requests.get(url)
data = r.json()
```

```python
df = pd.DataFrame(data['Data']['Data'])
df.head()
```

```python
maxTs = df.iloc[0]['time']
limit = 2000
histo = 'histohour'
oldestTs = 1388563200
count = 0
while maxTs >= oldestTs:
    maxTs = df.iloc[0]['time']
    
   
    maxTs -= 1
    # API call with the oldest ts
    url = f"https://min-api.cryptocompare.com/data/v2/{histo}?fsym=BTC&tsym=USD&limit={limit}&toTs={maxTs}&api_key={api_key}"
    r = requests.get(url)
    data = r.json()
    price_df = pd.DataFrame(data['Data']['Data'])
    df = price_df.append(df)
    count += 1
 
        
    
      
        


```
